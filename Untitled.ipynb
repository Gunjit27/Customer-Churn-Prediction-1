```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
df = pd.read_csv('customer_churn.csv')
df.sample(5)
```

```python
df = df.drop('customerID',axis='columns')
df.dtypes
```

```python
df.TotalCharges.values
```

```python
pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df[pd.to_numeric(df.TotalCharges,errors='coerce').isnull()]
```

```python
df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
df1.dtypes
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1.TotalCharges.dtypes
```

```python
def print_unique_col_values(df):
    for col in df:
        if df[col].dtype=='object':
            print(f'{col}: {df[col].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
yes_no_cols = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 
               'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

for col in yes_no_cols:
    df1[col].replace('Yes',1,inplace=True)
    df1[col].replace('No',0,inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('Female',1,inplace=True)
df1.replace('Male',0,inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
df2 = pd.get_dummies(data = df1, columns = ['InternetService' , 'Contract' , 'PaymentMethod'])
# df2 = df2.astype(int)
df2.sample(5)
```

```python
df2.replace(False,0,inplace=True)
df2.replace(True,1,inplace=True)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure' , 'MonthlyCharges' , 'TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

```python
x = df2.drop('Churn',axis='columns')
y = df2.Churn
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
x_train[:10]
```

```python
len(x_train.columns)
```

```python
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(26,)))
model.add(keras.layers.Dense(26,activation='relu'))
model.add(keras.layers.Dense(20,activation='relu'))
model.add(keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp = model.predict(x_test)
yp_int = []
for i in yp:
    if i>0.5:
        yp_int.append(1)
    else:
        yp_int.append(0)
yp_int[:10]
```

```python
y_test[:10]
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=yp_int)
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
